```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum, avg, max, min
```

```python
spark = SparkSession.builder.appName("BigDataAnalysis").getOrCreate()
```

```python
data = spark.read.csv("Sales Dataset.csv", header=True, inferSchema=True)
```

```python
data = data.dropna().dropDuplicates()
```

```python
data = data.withColumn("TotalSales", col("Amount") * col("Quantity"))
```

```python
print("Total Records:", data.count())
```

```python
print("Total Sales:")
data.select(sum("TotalSales")).show()

```

```python
print("Average Sales:")
data.select(avg("TotalSales")).show()
```

```python
print("Max Sales:")
data.select(max("TotalSales")).show()
```

```python
print("Sales by State:")
data.groupBy("State").sum("TotalSales").show()
```

```python
print("Sales by Category:")
data.groupBy("Category").sum("TotalSales").show()
```
